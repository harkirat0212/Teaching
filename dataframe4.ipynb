```python
df["Name"].astype(str).str.split(",")
```

```python
df['Name'].str.split(',').str[1]
```

```python
df["new_name"]=df['Name'].str.split(',').str[1]
```

```python
df["new_name"]=df['new_name'].str.strip()
df
```

```python
mr=df[df["new_name"].str.startswith("Mr.")]
mr
```

```python
mr.info()
```

```python
mr["Age"].isna()
```

```python
mr.Age.mean()
```

```python
mr.fillna(mr.Age.mean(),inplace=True)
mr.info()
```

```python
mrs=df[df["new_name"].str.startswith("Mrs.")]
mrs
```

```python
mrs.Age.mean()
```

```python
mrs.fillna(mrs.Age.mean(),inplace=True)
mrs.info()
```

```python
miss=df[df["new_name"].str.startswith("Miss")]
miss
```

```python
miss.Age.mean()
```

```python
miss.fillna(miss.Age.mean(),inplace=True)
miss.info()
```

```python
master=df[df["new_name"].str.startswith("Master.")]
master
```

```python
master.fillna(master.Age.mean(),inplace=True)
master.info()
```

```python
master.Age.mean()
```

```python
df=pd.concat([mr,mrs,master,miss])
```
